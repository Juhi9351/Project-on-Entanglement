```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

theta_vals = np.linspace(0, np.pi, 100)  # theta from 0 to pi
psi_vals = np.linspace(0, np.pi/2, 100)  # psi from 0 to pi/2

Theta, Psi = np.meshgrid(theta_vals, psi_vals)

F1 = 0.5 * (np.cos(Psi)**2 * np.cos(Theta) + np.sin(Psi)**2 * np.sin(Theta))**2

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(Theta, Psi, F1, cmap='viridis', edgecolor='none')

fig.colorbar(surf, shrink=0.5, aspect=10, label=r'$F_1$ Value')

ax.set_xlabel(r'$\theta$ (radians)')
ax.set_ylabel(r'$\psi$ (radians)')
ax.set_zlabel(r'$F_1$')
ax.set_title('3D Plot of Fidelity $F_1$')

ax.view_init(elev=30, azim=135)  # Adjust elevation & azimuth

ax.set_box_aspect([1, 1, 0.7]) 

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

theta_vals = np.linspace(0, np.pi, 100)  # theta from 0 to pi
psi_vals = np.linspace(0, np.pi/2, 100)  # psi from 0 to pi/2

Theta, Psi = np.meshgrid(theta_vals, psi_vals)

F1 = 0.5 * (np.cos(Psi)**2 * np.cos(Theta) + np.sin(Psi)**2 * np.sin(Theta))**2
F2 = 0.5 * (np.cos(Psi)**2 * np.sin(Theta) + np.sin(Psi)**2 * np.cos(Theta))**2

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(Theta, Psi, F1, cmap='viridis', edgecolor='none')
ax1.set_xlabel(r'$\theta$ (radians)')
ax1.set_ylabel(r'$\psi$ (radians)')
ax1.set_zlabel(r'$F_1$')
ax1.set_title('3D Plot of $F_1$')
ax1.view_init(elev=30, azim=135)
fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=10)

ax2 = fig.add_subplot(122, projection='3d')
surf2 = ax2.plot_surface(Theta, Psi, F2, cmap='plasma', edgecolor='none')
ax2.set_xlabel(r'$\theta$ (radians)')
ax2.set_ylabel(r'$\psi$ (radians)')
ax2.set_zlabel(r'$F_2$')
ax2.set_title('3D Plot of $F_2$')
ax2.view_init(elev=30, azim=135)
fig.colorbar(surf2, ax=ax2, shrink=0.5, aspect=10)

plt.show()

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

contour1 = ax[0].contourf(Theta, Psi, F1, cmap='viridis', levels=20)
ax[0].set_xlabel(r'$\theta$ (radians)')
ax[0].set_ylabel(r'$\psi$ (radians)')
ax[0].set_title('Contour Plot of $F_1$')
fig.colorbar(contour1, ax=ax[0])

contour2 = ax[1].contourf(Theta, Psi, F2, cmap='plasma', levels=20)
ax[1].set_xlabel(r'$\theta$ (radians)')
ax[1].set_ylabel(r'$\psi$ (radians)')
ax[1].set_title('Contour Plot of $F_2$')
fig.colorbar(contour2, ax=ax[1])

plt.show()
```
